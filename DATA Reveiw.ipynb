```python
# First cell - Setup and Data Loading
import os, types
import pandas as pd
import numpy as np
from botocore.client import Config
import ibm_boto3

# Helper function to read data from IBM COS
def load_cos_file(object_key, file_type='excel'):
    body = cos_client.get_object(Bucket=bucket, Key=object_key)['Body']
    if not hasattr(body, "__iter__"): 
        body.__iter__ = types.MethodType(__iter__, body)
        
    if file_type == 'excel':
        return pd.read_excel(body.read())
    else:
        return pd.read_csv(body)

# Load both datasets
crop_calendar = load_cos_file('Crop_Calendar_Data_All.xlsx', 'excel')
market_prices = load_cos_file('Prices-Export-Fri Oct 25 2024 05_50_02 GMT+0530 (India Standard Time).csv', 'csv')

# Second cell - Initial Data Analysis
# For Crop Calendar
print("Crop Calendar Data Info:")
print(crop_calendar.columns.tolist())
print("\nShape:", crop_calendar.shape)

# For Market Prices
print("\nMarket Prices Data Info:")
print(market_prices.columns.tolist())
print("\nShape:", market_prices.shape)
```

```python
# Improved Data Cleaning and Processing
def clean_crop_calendar(df):
    # Select and rename relevant columns
    cleaned_df = df[[
        'AgroEcological Zone', 
        'Crop',
        'Early Sowing',
        'Later Sowing',
        'Growing period',
        'Early harvest',
        'Late harvest',
        'AgroEcological Zone Description'
    ]].copy()
    
    # Remove rows where AgroEcological Zone is NaN
    cleaned_df = cleaned_df.dropna(subset=['AgroEcological Zone'])
    
    # Clean crop names
    cleaned_df['Crop'] = cleaned_df['Crop'].str.strip()
    
    return cleaned_df

def clean_market_prices(df):
    cleaned_df = df[[
        'Market Name',
        'Commodity',
        'Price Type',
        'Price Date',
        'Price',
        'Unit',
        'Currency',
        'ALPS Phase'
    ]].copy()
    
    # Convert price date to datetime with correct format
    cleaned_df['Price Date'] = pd.to_datetime(cleaned_df['Price Date'], dayfirst=True)
    
    # Convert price to numeric, removing any currency symbols
    cleaned_df['Price'] = pd.to_numeric(cleaned_df['Price'], errors='coerce')
    
    return cleaned_df

# Clean the data
clean_calendar = clean_crop_calendar(crop_calendar)
clean_prices = clean_market_prices(market_prices)

# Improved recommendation functions
def get_zone_info():
    """Get all available zones and their descriptions"""
    zones = clean_calendar[['AgroEcological Zone', 'AgroEcological Zone Description']].drop_duplicates()
    return zones.to_dict('records')

def get_crop_recommendations(zone):
    """Get crop recommendations for a specific agro-ecological zone"""
    zone_crops = clean_calendar[clean_calendar['AgroEcological Zone'] == zone]
    crops = zone_crops['Crop'].unique().tolist()
    return {
        'zone': zone,
        'recommended_crops': crops,
        'count': len(crops)
    }

def get_market_analysis(commodity, days_range=30):
    """Get market analysis for a specific commodity"""
    prices = clean_prices[clean_prices['Commodity'] == commodity]
    latest_date = prices['Price Date'].max()
    month_ago = latest_date - pd.Timedelta(days=days_range)
    
    recent_prices = prices[prices['Price Date'] >= month_ago]
    
    return {
        'commodity': commodity,
        'current_price': prices[prices['Price Date'] == latest_date]['Price'].mean(),
        'average_30day_price': recent_prices['Price'].mean(),
        'price_trend': prices['ALPS Phase'].mode().iloc[0],
        'currency': prices['Currency'].iloc[0],
        'markets': recent_prices['Market Name'].unique().tolist()
    }

# Test the improved functions
print("Available Zones and Descriptions:")
zones = get_zone_info()
for zone in zones[:2]:  # Show first two zones
    print(f"\nZone: {zone['AgroEcological Zone']}")
    print(f"Description: {zone['AgroEcological Zone Description'][:100]}...")  # First 100 chars

print("\nCrop Recommendations for Coastal Plains:")
print(get_crop_recommendations('Coastal Plains'))

print("\nMarket Analysis for Beans:")
print(get_market_analysis('Beans'))
```

```python
# Import weather api and existing modules
import os, types
import pandas as pd
import numpy as np
import requests
from datetime import datetime

# Weather API setup
WEATHER_API_KEY = 'YOUR_IBM_WEATHER_API_KEY'

class AgriBuddyAdvisor:
    def __init__(self, clean_calendar, clean_prices, weather_api_key):
        self.calendar = clean_calendar
        self.prices = clean_prices
        self.weather_api_key = weather_api_key
        
    def get_complete_advisory(self, location, zone):
        """
        Generate complete advisory for a location
        """
        weather = self._get_weather(location['latitude'], location['longitude'])
        crops = self.get_crop_recommendations(zone)
        market = self.get_market_info(crops['recommended_crops'])
        
        return {
            'crops': crops,
            'weather': weather,
            'market': market,
            'recommendations': self._generate_recommendations(crops, weather, market)
        }
    
    def get_crop_recommendations(self, zone):
        zone_data = clean_calendar[clean_calendar['AgroEcological Zone'] == zone]
        crops = zone_data['Crop'].unique().tolist()
        
        return {
            'zone': zone,
            'recommended_crops': crops,
            'total_crops': len(crops),
            'zone_description': zone_data['AgroEcological Zone Description'].iloc[0]
        }
    
    def _get_weather(self, lat, lon):
        """Get 7-day forecast from IBM Weather API"""
        url = f"https://api.weather.com/v1/geocode/{lat}/{lon}/forecast/daily/7day.json"
        params = {
            'apiKey': self.weather_api_key,
            'units': 'm',
            'language': 'en-US'
        }
        
        try:
            response = requests.get(url, params=params)
            if response.status_code == 200:
                return self._process_weather_data(response.json())
            return None
        except Exception as e:
            print(f"Weather API error: {e}")
            return None
    
    def get_market_info(self, crops):
        """Get market information for list of crops"""
        market_data = {}
        for crop in crops:
            try:
                prices = self.prices[self.prices['Commodity'].str.contains(crop, case=False)]
                if not prices.empty:
                    latest_date = prices['Price Date'].max()
                    latest_prices = prices[prices['Price Date'] == latest_date]
                    
                    market_data[crop] = {
                        'current_price': latest_prices['Price'].mean(),
                        'currency': latest_prices['Currency'].iloc[0],
                        'market_phase': latest_prices['ALPS Phase'].iloc[0],
                        'available_markets': latest_prices['Market Name'].unique().tolist()
                    }
            except:
                continue
        return market_data
    
    def _generate_recommendations(self, crops, weather, market):
        """Generate specific recommendations based on all data"""
        recommendations = []
        
        # Weather based recommendations
        if weather:
            if weather['rain_chance'] > 60:
                recommendations.append({
                    'type': 'weather',
                    'priority': 'high',
                    'advice': 'High chance of rain in next 7 days. Consider these activities:'
                    'actions': [
                        'Delay chemical spraying',
                        'Prepare drainage systems',
                        'Consider postponing planting of drought-loving crops'
                    ]
                })
        
        # Crop based recommendations
        for crop in crops['recommended_crops'][:5]:  # Top 5 crops
            if crop in market:
                price_info = market[crop]
                recommendations.append({
                    'type': 'crop',
                    'crop': crop,
                    'advice': f"Market status for {crop}: {price_info['market_phase']}",
                    'price': f"{price_info['current_price']} {price_info['currency']}",
                    'markets': f"Available in {len(price_info['available_markets'])} markets"
                })
        
        return recommendations

# Initialize the advisor
advisor = AgriBuddyAdvisor(clean_calendar, clean_prices, WEATHER_API_KEY)

# Test the system
test_location = {
    'latitude': -6.8235,  # Dar es Salaam
    'longitude': 39.2695
}

# Get advisory for Coastal Plains
advisory = advisor.get_complete_advisory(test_location, 'Coastal Plains')

# Print results
print("\nComplete Agricultural Advisory for Coastal Plains:")
print("\nRecommended Crops:")
print(f"Total crops suitable: {advisory['crops']['total_crops']}")
print("Top crops:", advisory['crops']['recommended_crops'][:5])

print("\nMarket Information:")
for crop, info in advisory['market'].items():
    print(f"\n{crop}:")
    print(f"Current price: {info['current_price']} {info['currency']}")
    print(f"Market phase: {info['market_phase']}")

if advisory.get('weather'):
    print("\nWeather Advisory:")
    print(advisory['weather'])

print("\nSpecific Recommendations:")
for rec in advisory['recommendations']:
    print(f"\n{rec['type'].upper()} RECOMMENDATION:")
    print(rec['advice'])
```

```python
class AgriBuddyAdvisor:
    def __init__(self, clean_calendar, clean_prices):
        self.calendar = clean_calendar
        self.prices = clean_prices
        
        # Add price trend analysis
        self.price_trends = self._calculate_price_trends()
        
    def _calculate_price_trends(self):
        """Calculate 30-day price trends for commodities"""
        trends = {}
        for commodity in self.prices['Commodity'].unique():
            commodity_data = self.prices[self.prices['Commodity'] == commodity]
            if not commodity_data.empty:
                latest_date = commodity_data['Price Date'].max()
                month_ago = latest_date - pd.Timedelta(days=30)
                
                recent_prices = commodity_data[commodity_data['Price Date'] >= month_ago]
                if not recent_prices.empty:
                    avg_price = recent_prices['Price'].mean()
                    trend_pct = ((recent_prices['Price'].iloc[-1] - recent_prices['Price'].iloc[0]) 
                                / recent_prices['Price'].iloc[0] * 100)
                    trends[commodity] = {
                        'average_price': avg_price,
                        'trend_percentage': trend_pct
                    }
        return trends

    def get_complete_advisory(self, zone):
        """Generate complete advisory for a zone"""
        crops = self.get_crop_recommendations(zone)
        market = self.get_market_info(crops['recommended_crops'])
        seasonal_info = self.get_seasonal_info(zone, crops['recommended_crops'])
        
        return {
            'zone_info': crops,
            'market_info': market,
            'seasonal_info': seasonal_info,
            'recommendations': self._generate_recommendations(crops, market, seasonal_info)
        }
    
    def get_seasonal_info(self, zone, crops):
        """Get planting and harvesting times for crops in the zone"""
        seasonal_data = {}
        zone_data = self.calendar[self.calendar['AgroEcological Zone'] == zone]
        
        for crop in crops:
            crop_data = zone_data[zone_data['Crop'] == crop]
            if not crop_data.empty:
                seasonal_data[crop] = {
                    'early_sowing': crop_data['Early Sowing'].iloc[0],
                    'later_sowing': crop_data['Later Sowing'].iloc[0],
                    'growing_period': crop_data['Growing period'].iloc[0],
                    'early_harvest': crop_data['Early harvest'].iloc[0],
                    'late_harvest': crop_data['Late harvest'].iloc[0]
                }
        return seasonal_data
    
    def _generate_recommendations(self, crops, market, seasonal_info):
        """Generate enhanced recommendations"""
        recommendations = []
        
        for crop in crops['recommended_crops']:
            rec = {
                'crop': crop,
                'market_info': {},
                'seasonal_info': {},
                'advice': []
            }
            
            # Add market information
            if crop in market:
                market_info = market[crop]
                rec['market_info'] = market_info
                
                # Add price trend analysis
                if crop in self.price_trends:
                    trend = self.price_trends[crop]
                    rec['advice'].append(self._get_price_trend_advice(crop, trend))
                
                # Add market phase advice
                rec['advice'].append(self._get_market_phase_advice(crop, market_info))
            
            # Add seasonal information
            if crop in seasonal_info:
                season = seasonal_info[crop]
                rec['seasonal_info'] = season
                rec['advice'].append(self._get_seasonal_advice(crop, season))
            
            recommendations.append(rec)
        
        return recommendations
    
    def _get_price_trend_advice(self, crop, trend):
        """Generate advice based on price trends"""
        if trend['trend_percentage'] > 10:
            return f"Price trending up significantly (+{trend['trend_percentage']:.1f}%). Good time to plan {crop} production."
        elif trend['trend_percentage'] < -10:
            return f"Price trending down (-{abs(trend['trend_percentage']):.1f}%). Consider storage or value addition for {crop}."
        else:
            return f"Stable price trends for {crop}. Regular production planning recommended."
    
    def _get_market_phase_advice(self, crop, market_info):
        """Generate detailed market phase advice"""
        phase = market_info['market_phase']
        price = market_info['current_price']
        currency = market_info['currency']
        
        advice = f"Current {crop} price: {price:,.0f} {currency}. "
        
        if phase == 'Crisis':
            advice += f"Market showing high stress. Consider immediate {crop} sales if available."
        elif phase == 'Alert':
            advice += f"Market showing increased demand. Good time to prepare {crop} production."
        elif phase == 'Stress':
            advice += f"Market showing some stress. Monitor {crop} prices closely."
        else:
            advice += f"Normal market conditions for {crop}."
            
        return advice
    
    def _get_seasonal_advice(self, crop, season):
        """Generate seasonal timing advice"""
        current_month = datetime.now().month
        
        advice = f"For {crop}: "
        advice += f"Early sowing period: {season['early_sowing']}, "
        advice += f"Later sowing period: {season['later_sowing']}, "
        advice += f"Growing period: {season['growing_period']}, "
        advice += f"Harvest period: {season['early_harvest']} to {season['late_harvest']}"
        
        return advice

# Test the enhanced system
advisor = AgriBuddyAdvisor(clean_calendar, clean_prices)

# Test for one zone to see detailed output
test_zone = 'Coastal Plains'
print(f"\nDetailed Advisory for {test_zone}")
print("=" * 50)

advisory = advisor.get_complete_advisory(test_zone)

print("\nRECOMMENDED CROPS:")
for crop in advisory['zone_info']['recommended_crops'][:5]:
    print(f"\n{crop.upper()}:")
    
    # Find recommendations for this crop
    crop_rec = next((r for r in advisory['recommendations'] if r['crop'] == crop), None)
    
    if crop_rec:
        print("\nAdvice:")
        for advice in crop_rec['advice']:
            print(f"- {advice}")
            
        if 'market_info' in crop_rec and crop_rec['market_info']:
            print(f"\nAvailable Markets: {', '.join(crop_rec['market_info']['available_markets'][:5])}")
```
